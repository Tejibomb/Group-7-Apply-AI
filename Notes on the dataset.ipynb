BankSim is based on a sample of agregated transaction data provided by one bank in Spain.

Data contains several thousand recordsof transactional data covering six months, from November2012 until April 2013

Limited to transactions in Barcelona and Madrid.


Data is grouped by month, week, day of the week and hour.


Three types of queries:

* consumption habits
* customer classification
* origin and source of transactions

provides information about payment like: 
* number
* average
* minimum values
* maximum values

also includes
* zipcode location of origin/source
* merchant category
* customer gender and age

* account name

# Cases of Fraud

**Theft**

occurs when physical card is stolen and the fraudster impersonates the owner to purchase goods or services with the stolen card.

a **red flag** for this  is could be a high number of unusual transactions with high value in a short period of time



**clone card/skimming**
a **red flag** for this could be a high number of unusual transactions with high value in a short period.

another one may be simultaenous payments in different physical locations 

or using the card far from previously known locations

**internet purchases**

 A ``red flag'' for detection inthis case could be to have a black list of Carding websitesand proceed to cross this information with current useractivity to detect any unusual purchases after the Cardingwas executed.
